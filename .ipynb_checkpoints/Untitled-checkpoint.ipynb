```python
with zipfile.ZipFile('train.csv.zip', 'r') as zip_file:
    file_names = zip_file.namelist()
    csv_file_name = file_names[0]
    with zip_file.open(csv_file_name) as csv_file:
        train = pd.read_csv(csv_file)
```
